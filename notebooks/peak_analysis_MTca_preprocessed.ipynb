```python
xlsx_path = "path/to/file.xlsx"  # <- update path
```

```python
col_index = 0  # column index to pull values from (0 = first column)

# read all sheets without headers so Excel row numbers map directly to df.iloc indices
sheets = pd.read_excel(xlsx_path, sheet_name=None, header=None)

dfs = []
for sheet_name, df in sheets.items():
    # select Excel rows 39-54 -> iloc[38:54] (stop is exclusive) and all columns
    sub = df.iloc[38:54, :].apply(pd.to_numeric, errors="coerce").reset_index(drop=True)
    if sub.shape[1] == 0:
        continue
    # name every column coming from this sheet with the sheet name
    sub.columns = [sheet_name] * sub.shape[1]
    dfs.append(sub)

# build final dataframe: one column per source column, each column labeled with its sheet name
if dfs:
    result_df = pd.concat(dfs, axis=1)
else:
    result_df = pd.DataFrame()

# optional: drop rows that are all NaN
result_df = result_df.dropna(how="all").reset_index(drop=True)

result_df
```
